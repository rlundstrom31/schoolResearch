```python
#old way to pull off strips
#     while not isEmpty(tabList):
#        #print(tabList, steps)
#         corner = upperRightNotEmpty(tabList)
#         i = corner[0]
#         j = corner[1]
#         goodPath = True
#         while goodPath:
#             if isInBounds(tabList,i, j-1) and tabList[i][j] == tabList[i][j-1]:
#                     tabList[i][j] -= 1
#                     j = j - 1
#             else:
#                 if not isInBounds(tabList, i+1, j):
#                     tabList[i][j] -= 1
#                     goodPath = False
#                     steps += 1
#                 else:
#                     tabList[i][j] -= 1
#                     i = i + 1

#methods
def isEmpty(tab):
    for i in range(len(tab)):
        for j in range(len(tab[i])-1,-1,-1):
            if tab[i][j] > 0:
                return False
    return True

def upperRightNotEmpty(tab):
    for i in range(len(tab)):
        for j in range(len(tab[i])-1,-1,-1):
            if tab[i][j] > 0:
                return [i,j]
    return [0,0]

def isInBounds(tab, i, j):
    if i >= 0 and j >= 0 and i < len(tab) and j < len(tab[i]):
        return True
    else:
        return False
def setup(tab):
    tabList = list()
    for i in range(len(tab)):
        rowList = list()
        for j in range(len(tab[i])):
            rowList.append(tab[i][j] - (i+1))
        tabList.append(rowList)
    return tabList

def fitsIn(this, that): #returns true if this "fits in" that - assume they have the same shape and check box-by-box
    for i in range(len(this)):
        for j in range(len(this[i])):
            if this[i][j] > that[i][j]:
                return False
    return True

def maxCorner(maxTab):
    best = 0
    for i in range(len(maxTab)):
        best = max(best, maxTab[i][len(maxTab[i])-1] + (i + 1))
    return best

def maxRemove(maxTab):
    runSum = 0
    for i in range(len(maxTab)):
        runSum += maxTab[i][-1]
    return runSum
def posSet(shapeHelp):
    shape = list()
    for x in shapeHelp:
        shape.append(x)
    shape.append(0)
    returnSet = set()
    for k in range(len(shape)-1):
        if (shape[k] != shape[k+1]):
            i = k
            j = shape[i]-1
            while(i >= 0 and j >= 0):
                returnSet.add(tuple([i,j]))
                i = i - 1
                j = j - 1
    i = len(shape)-1
    j = shape[i]
    return returnSet
def negSet(shape):
    returnSet = set()
    for k in range(len(shape)):
        if(k != 0):
            if (shape[k] != shape[k-1]):
                i = k-1
                j = shape[k]-1
                while(i >= 0 and j >= 0):
                    returnSet.add(tuple([i,j]))
                    i = i - 1
                    j = j - 1
    i = len(shape)-1
    j = shape[i]
    return returnSet
def sumOfEntries(lam):
    ans = 0
    for row in lam:
        for x in row:
            ans += x
    return ans
print("technologic")
```

```python
#crazy nQueens type recursive solution to making fdiagrams
def fDiagramBuilder(previous, i, j, ans):
    hookSum = 0
    for a in range(i):
        hookSum += previous[a][j]
    for b in range(j):
        hookSum += previous[i][b]
    col = j + 1
    maxPick = col - hookSum
    for k in range(maxPick + 1):
        tempNext = list()
        for a in range(len(previous)):
            tempRow = list()
            for b in range(len(previous[a])):
                tempRow.append(previous[a][b])
            tempNext.append(tempRow)
        tempNext[i][j] = k
        if isInBounds(previous, i, j+1):
            fDiagramBuilder(tempNext,i,j+1, ans)
        else:
            if isInBounds(previous,i+1,0):
                fDiagramBuilder(tempNext,i+1,0, ans)
            else:
                temp = list()
                for m in range(len(tempNext)):
                    temp.append(tuple(tempNext[m]))
                if sumOfEntries(temp) == 14:
                    ans.add(tuple(temp))
def maxPathLength(lam):
    startRow = 0
    startCol = len(lam[0]) - 1
    ans = list()
    for i in range(len(lam)):
        ans.append([-1] * len(lam[i]))
    sumAt(lam,startRow,startCol,ans)
    return ans
def sumAt(lam,i,j,ans):
    if i >= len(lam) or j < 0:
        return 0
    elif ans[i][j] != -1:
        return ans[i][j]
    else:
        ans[i][j] = lam[i][j] + max(sumAt(lam,i + 1,j,ans),sumAt(lam,i,j-1,ans))
        return ans[i][j]
```
