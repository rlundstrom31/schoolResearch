```python
#important functions
def sieve(n):
    test = [0,0]
    for i in range(2,n):
        test.append(i)
    for i in range(0, n):
        if (test[i] != 0):
            startNum = i*i
            while (startNum < n):
                test[startNum] = 0
                startNum = startNum+i
    primes = []
    for i in test:
        if (i != 0):
            primes.append(i)
    return primes
#have another version that does single numbers in eclipse...this one is better for a range of values
def factorize(n, primes, old):
    for i in primes:
        if (n%i == 0):
            found = old[int(n/i)]
            new = []
            for j in found:
                temp = []
                temp.append(j[0])
                temp.append(j[1])
                new.append(temp)
            
            if new is []:
                new.append([i,1])
            else:
                good = False
                for j in new:
                    if j[0] == i:
                        j[1] =j[1] + 1
                        good = True
                if not good:
                    new.append([i, 1])
    
    return new

def reverseArray(X):
    l = len(X)
    Y = [0]*l
    for i in range(l):
        Y[l-i-1] = X[i]
    return Y
```
