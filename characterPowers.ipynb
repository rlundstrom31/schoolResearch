```python
def partitionFunction(a,b,d):
    return Integer((a^4 - 10*a^3 + 27*a^2 - 18*a)/8) + Integer((-b^2 + 3*b)/2) - d + Integer((b*a*a - 3*b*a)/2)

def sizeFunction(a,b,d):
    n = a + 2*b + 4*d
    den = math.factorial(a) * math.factorial(b) * math.factorial(d) * 2^b * 4^d
    return Integer(math.factorial(n) / den)

def signFunction(a,b,d):
    return (-1)^(b+d)


thingsThatGoToIdentity = [[8,0,0], [6,1,0], [4,2,0], [2,3,0], [0,4,0], [4,0,1], [2,1,1], [0,2,1], [0,0,2]]

runningSum = 0
for thing in thingsThatGoToIdentity:
    thisFunction = partitionFunction(thing[0], thing[1], thing[2])
    thisSize = sizeFunction(thing[0],thing[1],thing[2])
    runningSum = runningSum + thisFunction*thisSize*signFunction(thing[0],thing[1],thing[2])
    print(thing, thisFunction, thisSize, signFunction(thing[0],thing[1],thing[2]))
    
print(runningSum)
```
